```python
import requests

def nve_api():
    url="https://biapi.nve.no/magasinstatistikk/api/Magasinstatistikk/HentOffentligDataMinMaxMedian"

    #Making the API request 
    response = requests.get(url)

    #Cheking if the request was successful
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Request failed with status code: {response.status_code}")
        return None

#Fetching and printing the data
data = nve_api()
if data:
    print(data)
```

```python
import requests
import matplotlib.pyplot as plt

def fetch_exchange_rates(api_url):
    response = requests.get(api_url)

    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Request failed with status code: {response.status_code}")
        return None

def visualize_exchange_rates(exchange_rates_data):
    dates = []
    exchange_rates = []

    for series in exchange_rates_data['data']['structure']['dimensions']['observation']:
        date = series['time_period']
        exchange_rate = float(series['values'][0]['value'])
        dates.append(date)
        exchange_rates.append(exchange_rate)

    plt.figure(figsize=(10, 6))
    plt.plot(dates, exchange_rates, marker='o')
    plt.xlabel('Date')
    plt.ylabel('Exchange Rate')
    plt.title('Exchange Rate Trends')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def main():
    api_url = "https://data.norges-bank.no/api/data/EXR/B.USD.NOK.SP?format=sdmx-json&startPeriod=2022-08-18&endPeriod=2023-08-18&locale=no"
    exchange_rates_data = exchange_rates(api_url)

    if exchange_rates_data:
        visualize_exchange_rates(exchange_rates_data)

data = exchange_rates(api_url)
if data:
    print(data)


    
```
